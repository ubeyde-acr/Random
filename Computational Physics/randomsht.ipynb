```python
def discrete_laplacian(M):
    """Get the discrete Laplacian of matrix M"""
    L = -4*M
    L += np.roll(M, (0,-1), (0,1)) # right neighbor
    L += np.roll(M, (0,+1), (0,1)) # left neighbor
    L += np.roll(M, (-1,0), (0,1)) # top neighbor
    L += np.roll(M, (+1,0), (0,1)) # bottom neighbor

    return L
```

```python
discrete_laplacian(A)
```

```python
def gray_scott_update(A, B, DA, DB, f, k, delta_t):
    """
    Updates a concentration configuration according to a Gray-Scott model
    with diffusion coefficients DA and DB, as well as feed rate f and
    kill rate k.
    """

    # Let's get the discrete Laplacians first
    LA = discrete_laplacian(A)
    LB = discrete_laplacian(B)

    # Now apply the update formula
    diff_A = (DA*LA - A*B**2 + f*(1-A)) * delta_t
    diff_B = (DB*LB + A*B**2 - (k+f)*B) * delta_t

    A += diff_A
    B += diff_B

    return A, B
```

```python
def get_initial_configuration(N, random_influence=0.2):
    """
    Initialize a concentration configuration. N is the side length
    of the (N x N)-sized grid.
    `random_influence` describes how much noise is added.
    """

    # We start with a configuration where on every grid cell
    # there's a lot of chemical A, so the concentration is high
    A = (1-random_influence) * np.ones((N,N)) + random_influence * np.random.random((N,N))

    # Let's assume there's only a bit of B everywhere
    B = random_influence * np.random.random((N,N))

    # Now let's add a disturbance in the center
    N2 = N//2
    radius = r = int(N/10.0)

    A[N2-r:N2+r, N2-r:N2+r] = 0.50
    B[N2-r:N2+r, N2-r:N2+r] = 0.25

    return A, B
```

```python
def draw(A,B):
    """draw the concentrations"""
    fig, ax = plt.subplots(1,2,figsize=(5.65,4))
    ax[0].imshow(A, cmap='Greys')
    ax[1].imshow(B, cmap='Greys')
    ax[0].set_title('A')
    ax[1].set_title('B')
    ax[0].axis('off')
    ax[1].axis('off')
```

```python
import matplotlib.animation as animation  # this is needed for the animation API
from matplotlib.colors import Normalize   # this is needed to rescale the color during the simulation
```

```python
def get_artists(A,B):
    """draw the concentrations"""
    fig, ax = plt.subplots(1,2,figsize=(5.65,3))
    imA = ax[0].imshow(A, animated=True, cmap='Greys')
    imB = ax[1].imshow(B, animated=True, cmap='Greys')
    ax[0].set_title('A')
    ax[1].set_title('B')
    ax[0].axis('off')
    ax[1].axis('off')

    return fig, imA, imB
```

```python
def updatefig(frame_id,updates_per_frame,A,B,DA,DB,f,k,delta_t):
    """Takes care of the matplotlib-artist update in the animation"""
    global imA, imB

    # update x times before updating the frame
    for u in range(updates_per_frame):
        A, B = gray_scott_update(A,B,DA,DB,f,k,delta_t)

    # update the frame
    imA.set_array(A)
    imB.set_array(B)

    # renormalize the colors
    imA.set_norm(Normalize(vmin=np.amin(A),vmax=np.amax(A)))
    imB.set_norm(Normalize(vmin=np.amin(B),vmax=np.amax(B)))

    # return the updated matplotlib objects
    return imA, imB
```

```python
# Diffusion coefficients
DA = 0.16
DB = 0.08

# define feed/kill rates
f = 0.055
k = 0.062

N_simulation_steps = 10000

# intialize the figures
A, B = get_initial_configuration(N)
fig, imA, imB = get_artists(A, B)

# how many updates should be computed before a new frame is drawn
updates_per_frame = 10

# these are the arguments which have to passed to the update function
animation_arguments = (updates_per_frame, A, B, DA, DB, f, k, delta_t)

# let's make sure we can see the animation in the notebook
from IPython.display import HTML
mpl.rcParams['animation.html'] = 'html5'
#mpl.rcParams['animation.embed_limit'] = 1024 # this allows jupyter notebook to save 1024MB video data in the memory


# start the animation
ani = animation.FuncAnimation(fig, #matplotlib figure
                              updatefig, # function that takes care of the update
                              fargs=animation_arguments, # arguments to pass to this function
                              interval=10, # update every `interval` milliseconds
                              frames=int(N_simulation_steps/updates_per_frame), #limit the number of frames
                              blit=True, # optimize the drawing update
                              )

HTML(ani.to_html5_video())
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
y0 = [0,0] #initial condition, 0 mRNA, 0 protein

t = np.linspace(0,200,num=200)

k_m = 0.2
gamma_m = 0.05
k_p = 0.4
gamma_p = 0.1

params = [k_m,gamma_m,k_p,gamma_p]
```

```python
def sim(variables,t,params):
  m = variables[0]
  p = variables[1]

  k_m =params[0]
  gamma_m = params[1]
  k_p = params[2]
  gamma_p = params[3]

  dmdt = k_m - gamma_m*m
  dpdt = k_p*m - gamma_p*p

  return([dmdt,dpdt])
```

```python
y = odeint(sim, y0, t, args=(params,))
```

```python
def sim2(variables,t,params):
  G1 = variables[0]
  G2 = variables[1]

  k1 =params[0]
  y1 = params[1]
  k2 = params[2]
  y2 = params[3]
  n=params[4]
  c=params[5]

  dG1dt = k1 - y1*G1
  dG2dt = ((G1**n)/(c**n + G1**n))*k2 - y2*G2

  return([dG1dt,dG2dt])
```

```python
import random

X=[0]

t=[0]

tend = 1000

k = 2
gamma = 0.1

while t[-1] < tend:
  current_X = X[-1]
  rates = [k,gamma*current_X]
  rate_sum = sum(rates)
  tau = np.random.exponential(scale=1/rate_sum)
  t.append(t[-1] + tau)
  rand = random.uniform(0,1)
  #Production Event
  if rand*rate_sum > 0 and rand * rate_sum < rates[0]:
    X.append(X[-1] + 1)
  #Decay Event
  if rand*rate_sum > rates[0] and rand * rate_sum < rates[0] + rates[1]:
    X.append(X[-1] - 1)

plt.plot(t,X)
plt.xlabel("time")
plt.ylabel("mRNA quantity")
plt.show()


```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python

# Create a figure to display the predator-prey equations
fig, ax = plt.subplots(figsize=(8, 4))  # Reduced figure size
ax.axis('off')

# Define the Lotka-Volterra equations
equations = r"""
$\frac{dx}{dt} = \alpha x - \beta xy$

$\frac{dy}{dt} = \delta xy - \gamma y$

Where:
x: Prey population
y: Predator population
$\alpha$: Prey growth rate
$\beta$: Prey death rate due to predation
$\delta$: Predator growth rate from predation
$\gamma$: Predator death rate
"""

# Display the equations
ax.text(0.5, 0.5, equations, fontsize=12, ha='center', va='center', 
        bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

plt.title("Lotka-Volterra Predator-Prey Model Equations", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
y0 = [10,1] #[Fish,Bears]


t = np.linspace(0,50,num=1000)

a = 1.5  # prey growth rate
b = 0.9  # prey death rate
d = 0.3  # predator growth rate
g = 0.6  # predator death rate

y1 = [g/d,a/b] #steady state condition

params = [a,b,d,g]
```

```python
def sim(variables,t,params):
    F = variables[0]
    B = variables[1]

    a = params[0]
    b = params[1]
    d = params[2]
    g = params[3]

    dxdt = a*F - b*F*B
    dydt = d*F*B - g*B

    return ([dxdt,dydt])
```

```python
pp = odeint(sim,y1,t,args=(params,))

f, ax = plt.subplots(1,figsize=(10,6))
line1, = ax.plot(t,pp[:,0], color="b",label="Fish")
line2, = ax.plot(t,pp[:,1], color="r",label="Bear")
ax.set_ylabel('Number')
ax.set_xlabel('Time')
ax.legend(handles=[line1,line2])
plt.show()

```

```python
import random

X=[1]

t=[0]

tend = 500

r = 0.04
K = 100

while t[-1] < tend:
  current_X = X[-1]
  rates = [r*current_X,r*((current_X**2)/K)]
  rate_sum = sum(rates)
  tau = np.random.exponential(scale=1/rate_sum)
  t.append(t[-1] + tau)
  rand = random.uniform(0,1)
  #Growth Event
  if rand*rate_sum > 0 and rand * rate_sum < rates[0]:
    X.append(X[-1] + 1)
  #Death Event
  if rand*rate_sum > rates[0] and rand * rate_sum < rates[0] + rates[1]:
    X.append(X[-1] - 1)

plt.plot(t,X,color="b",label="Deterministic")
plt.plot(time,log[:,0],color="r",label="Stochastic")
plt.xlabel("time")
plt.ylabel("Growth")
plt.legend()
plt.show()
```

```python
from rdkit import Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw
import matplotlib.pyplot as plt

# Define the polypyrrole monomer unit as a SMILES string
polyethylene_monomer_smiles = "CC"  # Pyrrole: C4H5N (aromatic ring with NH)

# Define the number of monomer units in the polymer chain
n_units = 30

# Function to create a polymer sequence
def create_polymer(monomer, n_units):
    return f"{'CC' * n_units}"  # Simplified SMILES for polypyrrole

# Create the polymer sequence (pyrrole repeating units)
polymer_smiles = create_polymer(polyethylene_monomer_smiles, n_units)
print("Polymer SMILES: ", polymer_smiles)

# Convert the polymer SMILES to an RDKit molecule object
polymer_molecule = Chem.MolFromSmiles(polymer_smiles)

# Generate a 3D structure
AllChem.EmbedMolecule(polymer_molecule, randomSeed=42)
AllChem.UFFOptimizeMolecule(polymer_molecule)

# Output the structure as a PNG image
img = Draw.MolToImage(polymer_molecule, size=(400, 400))

# Display the image using matplotlib
plt.figure(figsize=(5, 5))
plt.imshow(img)
plt.axis('off')
plt.title('Polyethylene Structure')
plt.show()
```
