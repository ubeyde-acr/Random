# Basics

The main purpose of NumPy is to provide an object, the `numpy.array` that allows for fast mathematical operations on multiple numbers at the same time without the use of forloops.

```python
x = np.array([1,3,5,6])
```

```python
x = np.array([1,3,5,6])
x**2
```

```python
np.exp(np.sin(x))
```

For plotting, it can be useful to create arrays of evenly spaced data. There are two ways to do this

```python
x = np.linspace(0,1,101)
x = np.arange(0,1,0.01)
```

Then you can apply a mathematical operation to this data

```python
y = np.sin(np.pi * x)
```

Or you can define your own function, and feed in a numpy array as one of the arguments:

```python
def f(x,a,b):
    return a*x**2 + b
```

```python
y = f(x, 2, 1)
```

And create a basic plot

```python
plt.plot(x,y)
```

Or I can add stuff to the plot to make it look fancier:

```python
plt.figure(figsize=(6,3))
plt.plot(x,y,'o--')
plt.xlabel('x')
plt.ylabel('y')
plt.title('x vs y')
plt.grid()
```

You can index particular elements of the arrays we just created:

```python
y[3]
```

```python
y[3]
```

# Multidimensional Arrays

Arrays can also be multidimensional:

```python
l = [[1,2,3],[10,44,55]]
```

```python
x = np.array([[1,2,3],[10,44,55]])
x
```

Whenever you deal with multidimensional arrays, you should always consider the shape:

```python
x.shape
```

This says that the array is 2 dimensional, and the first dimension has two elements (i.e. two rows) and the second dimension has 3 elements (3 columns)

* Indexing like this gives you rows:

```python
x[0]
```

* Indexing like this gives you columns:

```python
x[:,1]
```

You can index twice to get a particular element:

```python
x[0,1]
```

You can transpose a matrix, swapping the rows and columns

```python
x.T
```

Indexing the rows of the transposed matrix is the same as indexing the columns of the regular matrix

```python
x
```

```python
x[:,1]
```

```python
x.T[1]
```

# Boolean Operations

Some of the most useful features of the numpy array is boolean indexing. For example, suppose I create an array of evenly spaced values between 1 and 4

```python
x = np.linspace(1, 4, 100)
y = np.sin(x)
```

```python
plt.plot(x,y)
```

The following evaluates $x>2$ for all points in the `x` array

```python
x
```

```python
x>2
```

I can then use this array to index elements of $x$

```python
y[x>2]
```

I can even use the `x>2` array to index `y` and plot a subset of each array

```python
plt.plot(x,y)
plt.plot(x[y<0.5], y[y<0.5])
```

You can also combine multiple boolean conditions together:

```python
arr =(x>2) * (x<3)
```

You can think of `True`s as being 1s and `False`s as being 0s; thats how the mathematical operations are achieved on boolean values. Finally you can index with these as well

```python
plt.plot(x,y)
plt.plot(x[(y>-0.5) * (y<0.5)], y[(y>-0.5) * (y<0.5)])
```

* Most data is stored in `csv` files, which is essentially just a text file with a **delimiter** to seperate between data entries, and a line break to seperate between rows. The delimter is typically either a blank space or a comma
